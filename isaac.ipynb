```python
import numpy as np
import matplotlib.pyplot as plt
```

# Ex 6

```python
def f(x0, t):
  if x0 < -1:
    return x0+27*t
  return x0+12*t

def choque(t):
  return -1+19*t

t = np.linspace(0,2,100)
x0 = np.concatenate((np.linspace(-20,-1,20), np.linspace(-0,20,21)))

xs = choque(t)

fig, (a1, a2) = plt.subplots(1,2)
for i in x0:
  x = f(i,t)
  a1.plot(x,t)
  for j in range(len(t)):
    if i < -1 and x[j] > xs[j]:
      x[j] = None
    if i >= -1 and x[j] < xs[j]:
      x[j] = None
  a2.plot(x,t)
plt.savefig('ex6')
```

```python
def f(x0, t):
  if x0 < 1:
    return x0+3*t
  return x0+12*t

def rarefacao(x0,t,c):
  return x0+c*t

t = np.linspace(0,2,100)
x0 = np.concatenate((np.linspace(-20,1,22), np.linspace(1,20,20)))

c = np.linspace(3,12, 10)

fig, (a1, a2) = plt.subplots(1,2)
for i in x0:
  x = f(i,t)
  a1.plot(x,t)
  a2.plot(x,t)

for _c in c:
  x = rarefacao(1,t,_c)
  a2.plot(x,t)



plt.savefig('ex7')
```

# Ex 8

```python
def f(x0, t):
  if x0>0 and x0<1:
    return x0+4*t
  return x0+2*t

def rarefacao(x0,t,c):
  return x0+c*t

def choque(t):
  return 1+3*t

t = np.linspace(0,2,100)
x0 = np.linspace(-5,5,42)
c = np.linspace(2,4, 10)
xs = choque(t)

fig, (a1, a2) = plt.subplots(1,2)
for i in x0:
  x = f(i,t)
  a1.plot(x,t)
  for j in range(len(t)):
    if i < 1 and x[j] > xs[j]:
      x[j] = None
    if i >= 1 and x[j] < xs[j]:
      x[j] = None
  a2.plot(x,t)

for _c in c:
  x = rarefacao(0,t,_c)
  a2.plot(x,t)

plt.savefig('ex8')
```
